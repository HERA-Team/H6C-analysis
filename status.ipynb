```python
numbers = {'good': [], 'suspect': [], 'bad': []}
combined = {}

for k, v in statuses.items():
    if v is None:
        continue

    bad = set(v['ee']['bad'] + v['nn']['bad'])
    suspect = set(v['ee']['suspect'] + v['nn']['suspect'])
    suspect = [s for s in suspect if s not in bad]
    good = set(v['ee']['good'] + v['nn']['good'])
    good = [g for g in good if g not in suspect and g not in bad]
        
    combined[k] = {'good': good, 'suspect': suspect, 'bad': bad}
    numbers['good'].append(len(good))
    numbers['suspect'].append(len(suspect))
    numbers['bad'].append(len(bad))
```

```python

labels = [k for k, v in statuses.items() if v is not None]
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]
men_std = [2, 3, 4, 1, 2]
women_std = [3, 5, 2, 3, 3]
width = 0.35       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, numbers['good'], 0.8, label='Good', color='green')
ax.bar(labels, numbers['suspect'], 0.8, label='Suspect', bottom=numbers['good'], color='orange')
ax.bar(labels, numbers['bad'], 0.8, label='Bad', bottom=np.array(numbers['good']) + np.array(numbers['suspect']), color='red')

ax.set_ylabel('Number of Antennas')
ax.legend();
```

```python
from py21cmsense.antpos import hera
```

```python
antpos = hera(hex_num=11, split_core=True, outriggers=2)
```

```python
len(antpos)
```

```python
scores = []
notcon = []
for i in range(len(antpos)):
    score = 0
    not_connected = True
    for v in combined.values():
        if i in v['good']:
            score += 1
            not_connected = False
        elif i in v['suspect']:
            score += 0.5
            not_connected = False
        elif i in v['bad']:
            not_connected = False
            
    scores.append(score)
    if not_connected:
        notcon.append(i)
```

```python
mask = np.array([i not in notcon for i in range(len(antpos))])

plt.figure(figsize=(12,12))
plt.scatter(antpos[mask, 0], antpos[mask, 1], c=np.array(scores)[mask], cmap="RdYlGn")
plt.xlim(-150,150)
plt.ylim(-150,150)
clb = plt.colorbar()
clb.set_label("# Good Nights")
```

```python
matrix_version = np.zeros((len(antpos), len(combined)))

for j, (day, status) in enumerate(combined.items()):
    for i in range(len(antpos)):
        if i in status['good']:
            matrix_version[i,j] = 2
        elif i in status['suspect']:
            matrix_version[i,j] = 1
        elif i in status['bad']:
            matrix_version[i,j] = 0
        else:
            matrix_version[i,j] = np.nan
```
